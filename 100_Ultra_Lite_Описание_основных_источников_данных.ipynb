Если первый алгоритм на обработку 10 записей затрачивает одну операцию, а второй алгоритм на обработку 1 записи также затрачивает одну операцию, то оба алгоритма работают с одинаковой скоростью.

Алгоритмическая сложность первого алгоритма: O(1) - константная сложность.
Алгоритмическая сложность второго алгоритма: O(1) - также константная сложность.

Оба алгоритма имеют постоянную сложность, что означает, что время выполнения не зависит от размера входных данных.

# 3

Для сравнения скорости работы алгоритмов нужно учесть не только количество записей, но и количество операций, затрачиваемых на их обработку.

Первый алгоритм на обработку 1000 записей затрачивает 3 операции, а второй алгоритм на обработку 1000000 записей затрачивает 6 операций.

Алгоритмическая сложность первого алгоритма: O(3) - также константная сложность.
Алгоритмическая сложность второго алгоритма: O(6) - также константная сложность.

Таким образом, сравнивая количество операций, можно сделать вывод, что первый алгоритм работает быстрее, так как он затрачивает меньшее количество операций на обработку данных.
