```python
import gtts
from gtts import gTTS
from IPython.display import Audio
```

# Синтез аудио из текста

```python
# Словарь со всеми доступными языками
# К сожалению, для русского языка доступен только один голос
gtts.lang.tts_langs()
```

```python
# На русском языке синтез пословицы
text = """Многие знают много, но никто не знает всё."""

tts = gTTS(text, lang="ru") # преобразовываем текст в аудио
tts.save("proverb_ru.mp3") # сохраняем в формате mp3
```

```python
Audio("proverb_ru.mp3")
```

```python
# На английском языке синтез пословицы
text = """Многие знают много, но никто не знает всё."""

tts = gTTS(text, lang="en") # преобразовываем текст в аудио
tts.save("proverb_en.mp3") # сохраняем в формате mp3
```

```python
Audio("proverb_en.mp3")
```

```python
# На украинском языке синтез пословицы
text = """Многие знают много, но никто не знает всё."""

tts = gTTS(text, lang="uk") # преобразовываем текст в аудио
tts.save("proverb_uk.mp3") # сохраняем в формате mp3
```

```python
Audio("proverb_uk.mp3")
```
