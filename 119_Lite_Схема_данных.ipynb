```python
CREATE TABLE "public.Customers" (
	"id" serial NOT NULL UNIQUE,
	"name" varchar(255) NOT NULL,
	"industry" varchar(255) NOT NULL,
	"project1_id" integer(6) NOT NULL,
	"project1_feedback" TEXT NOT NULL,
	"project2_id" integer(6) NOT NULL,
	"project2_feedback" TEXT NOT NULL,
	"contact_person_id" integer NOT NULL,
	"contact_person_and_role" varchar NOT NULL,
	"phone_number" varchar NOT NULL,
	"address" varchar NOT NULL,
	"city" varchar NOT NULL,
	"zip" varchar NOT NULL,
	"project_id" integer(6) NOT NULL,
	CONSTRAINT "Customers_pk" PRIMARY KEY ("id")
) WITH (
  OIDS=FALSE
);



CREATE TABLE "public.projects" (
	"id" serial(6) NOT NULL,
	"name" varchar NOT NULL,
	"value" float4(8) NOT NULL DEFAULT '1',
	"start_date" DATE(8) NOT NULL,
	"end_date" DATE(8) NOT NULL,
	CONSTRAINT "projects_pk" PRIMARY KEY ("id")
) WITH (
  OIDS=FALSE
);



CREATE TABLE "public.job_orders" (
	"id" serial(6) NOT NULL,
	"employee_id" integer(6) NOT NULL,
	"project_id" integer(6) NOT NULL,
	"description" varchar(300) NOT NULL,
	"order_date_time" TIMESTAMP NOT NULL,
	"quantity" integer(6) NOT NULL,
	"price" float4(6) NOT NULL DEFAULT '2',
	CONSTRAINT "job_orders_pk" PRIMARY KEY ("id")
) WITH (
  OIDS=FALSE
);



CREATE TABLE "public.project_employees" (
	"id" serial NOT NULL,
	"project_id" integer NOT NULL,
	"employee_id" integer NOT NULL,
	"hours" float4 NOT NULL,
	CONSTRAINT "project_employees_pk" PRIMARY KEY ("id")
) WITH (
  OIDS=FALSE
);



CREATE TABLE "public.employees" (
	"id" serial NOT NULL,
	"first_name" varchar(300) NOT NULL,
	"last_name" varchar(300) NOT NULL,
	"hourly_wage" float4(5) NOT NULL DEFAULT '2',
	"hire_date" DATE NOT NULL,
	CONSTRAINT "employees_pk" PRIMARY KEY ("id")
) WITH (
  OIDS=FALSE
);



ALTER TABLE "Customers" ADD CONSTRAINT "Customers_fk0" FOREIGN KEY ("project_id") REFERENCES "projects"("id");


ALTER TABLE "job_orders" ADD CONSTRAINT "job_orders_fk0" FOREIGN KEY ("employee_id") REFERENCES "employees"("id");
ALTER TABLE "job_orders" ADD CONSTRAINT "job_orders_fk1" FOREIGN KEY ("project_id") REFERENCES "projects"("id");

ALTER TABLE "project_employees" ADD CONSTRAINT "project_employees_fk0" FOREIGN KEY ("project_id") REFERENCES "projects"("id");
ALTER TABLE "project_employees" ADD CONSTRAINT "project_employees_fk1" FOREIGN KEY ("employee_id") REFERENCES "employees"("id");

```
