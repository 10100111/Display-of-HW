У нас B2B проект "Мой Агент" по продаже авиаперевозок. Поставщик - агрегатор авиакомпаний. Клиенты -туристические агенты регистрируются в ЛК и ведут продажи авиабилетов для своих туристов.

Витрина для отдела кадров:

- Фонд оплаты труда
  - Цель: определить расходы по оплате труда
  - Агрегация: ФОТ = сумма зарплат по всем сотрудникам
- департаменты ФОТ
  - Цель: отчет по зарплатам всех сотрудников каждого департамента
  - Агрегация: сумма зарплат каждого департамента
- новые сотрудники
  - Цель: определить количество новых сотрудников
  - Агрегация: сумма сотрудников нанятых за период
- все сотрудники
  - Цель: определить сколько сотрудников работает в компании
  - Агрегация: сумма всех сотрудников компании

Витрина для отдела маркетинга:

- Количество заключенных договоров
  - Цель: определить сколько договоров заключено с клиентами
  - Агрегация: сумма договоров в статусе "подписан"
- Количество агентов, начавших работать
  - Цель: определить клиентов, которые начали оформлять заказы
  - Агрегация: список клиентов со статусом "оформлен первый заказ"
- Распределение договоров по кураторам
  - Цель: определить количество договоров у куратора
  - Агрегация: сгруппировать договоры по кураторам и посчитать сумму
- Количество заключенных договоров в городе
  - Цель: определить количество договоров в конкретном городе
  - Агрегация: сгруппировать договоры со статусом "подписан" по условию город
- Количество агентов, начавших работать в городе
  - Цель: определить сколько клиентов начали работать
  - Агрегация: список клиентов со статусом "оформлен первый заказ" сгруппировать по городу
- Статус клиента
  - Цель: получить список клиентов с нужным статусом
  - Агрегация: сгруппировать клиентов по конкретному статусу


Продажи:

- Месяц
  - Цель: определить сколько заказов оформлено за месяц
  - Агрегация: сумма всех заказов со статусом "продажа" за месяц + вывести общую стоимость

- Город
  - Цель: определить сколько заказов оформлено за месяц в городе
  - Агрегация: всех заказов со статусом "продажа" за месяц по условию город + вывести общую стоимость total
- Все продажи
  - Цель: определить сумму всех продаж и дохода
  - Агрегация: вывести сумму всех заказов по колонкам Тариф/fare, Такса/tax, Сборы/fee
- По Клиентам
  - Цель: определить объемы продаж по клиентам
  - Агрегация: сумма всех заказов по каждому customer + вывести общую стоимость total


Рейтинг:
- Агентство
  - Цель: определить лидеров продаж
  - Агрегация: Рейтинг = сортированный по убыванию список (сумма всех продаж клиента по количеству заказов) за период

Статус заказа:
- продажа
  - Цель: определить сколько заказов всего продано
  - Агрегация: сумма всех заказы за период со статусом "продажа"
- возврат
  - Цель: определить сколько заказов всего возвращено
  - Агрегация: сумма всех заказы за период со статусом "возврат"
