```python
import gtts
from gtts import gTTS
from IPython.display import Audio
```

# Синтез аудио из текста

```python
# Словарь со всеми доступными языками
# К сожалению, для русского языка доступен только один голос
gtts.lang.tts_langs()
```

```python
# Записываем диалог, как в документации
tts_ru = gTTS('Добрый день!', lang='ru')
tts_uk = gTTS('Добрый день!', lang='uk')
tts_ru1 = gTTS('Сколько стоят арбузы?', lang='ru')
tts_uk1 = gTTS('50 рублей за килограмм.', lang='uk')
tts_ru2 = gTTS('Мне, пожалуйста, самый маленький', lang='ru')
tts_uk2 = gTTS('Вот, держите.', lang='uk')
tts_ru3 = gTTS('Спасибо. До свидания.', lang='ru')
tts_uk3 = gTTS('До свидания.', lang='uk')

with open('dialog.mp3', 'wb') as f:
    tts_ru.write_to_fp(f)
    tts_uk.write_to_fp(f)
    tts_ru1.write_to_fp(f)
    tts_uk1.write_to_fp(f)
    tts_ru2.write_to_fp(f)
    tts_uk2.write_to_fp(f)
    tts_ru3.write_to_fp(f)
    tts_uk3.write_to_fp(f)
Audio('dialog.mp3')
```
