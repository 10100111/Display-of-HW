# Синтез аудио из текста

```python
text = """Ария кошек
Наши щёчки волосаты.
Наши спинки полосаты,
словно нотные листы.
Лапки — чудо красоты!

Красоты мы необычной,
выгнут хвост, как ключ скрипичный.
Мы в пыли его влачим
и в молчании — звучим."""

tts = gTTS(text, lang="ru") # преобразовываем текст в аудио
tts.save("aria_of_cats.mp3") # сохраняем в формате mp3
```

```python
Audio("aria_of_cats.mp3")
```
